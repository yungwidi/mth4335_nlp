## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.dpi"] = 120
%matplotlib inline
```

## Loading Sample

```python
df = pd.read_parquet("../ext/sample_50k.parquet")

#word counts
df["title_wc"]    = df["title"].str.split().apply(len)
df["abstract_wc"] = df["abstract"].str.split().apply(len)
df["title_cc"]    = df["title"].str.len()
df["abstract_cc"] = df["abstract"].str.len()

print(f"Loaded: {len(df):,} rows")
df.head()
```

```python
df["num_categories"] = df["categories"].str.split().apply(len)

print("Category count per paper:")
print(df["num_categories"].value_counts().sort_index().to_string())

multi_pct = (df["num_categories"] > 1).mean() * 100
print("")
print(f"Papers w/ multiple categories: {multi_pct:.1f}%")
```
